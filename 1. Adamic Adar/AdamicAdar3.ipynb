```python
df = pd.read_csv('Corrupted1_20_large.csv')
```

```python
df
```

```python
mean=df['Mystery_Data_Y'].mean(axis=0)
range=mean/4
```

```python
range
```

```python
df['Adamic_Adar_index']=0
```

```python
indexs=[]
```

```python
for i in df['Mystery_Data_Y']:
  first=1
  second=1
  third=1
  for j in df['Mystery_Data_Y']:
    if(abs(j-i)<=range):
      first=first+1
    if(abs(j-i)<=2*range):
      second=second+1
    if(abs(j-i)<=3*range):
      third=third+1
  index=0
  if(first!=1):
    index=(1/math.log(first))
  if(second!=1):
    index=index+(1/math.log(second))
  if(third!=1):
    index=index+(1/math.log(third))
  indexs.append(index)
```

```python
df = pd.DataFrame(indexs)
```

```python
df
```

```python
df.to_csv('AA.csv')
```
